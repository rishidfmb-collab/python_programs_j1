# operators in python

## python have 6 types of operators
1. arthematic operators
2. assignment operators
3. compartion operators
4. logical operators
5. membership operators
6. identity operators 

```python
#arthematic operators-+,-,*,/,//,%
a=10
b=20
c=30
d=40
print(a+b)
print(b-d)
print(a*c)
print(b/a)
print(b//a)
print(d%b)
```

```python
#comparision operators - <,>,>=,==,!= (return True or False)
a=5
b=10
c=10
d=20
print(a<b)
print(a>b)
print(b>=c)
print(b==c)
print(b!=d)
```

```python
#logical operators - and ,or, not
#and -this operators checking all conditions are ture result is ture rest cases failed
#or - this operators check either one conditions true true then result true rest cases failed
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b or b!=c)
print(not a>b or b!=c)
```

```python
# membership operators - checking members are present in that sequence or collection - in, not in
names=["shiva", "vikas","chandar"]
print("vikas" in names)
print("nani" in names)
```

```python
# identity operators - memory identification of values - is,is not
a=10000
b=10000
print(id(a))
print(id(b))
print(a is b)
```

```python
a=10
b=10
print(id(a))
print(b is not a)
```
