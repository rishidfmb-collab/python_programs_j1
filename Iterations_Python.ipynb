```python - for loop
for iter_variable in sequence:
    print("for is keyword to start the iterations")
    print("iter_variable to iterate data from sequence")
    print("in is keyword to validate the sequence is present or not")
```

```python - while loop
while (condition):
    print("while is keyword to start the loop and validation conditions based on values")
    print("if condition is true then inside statements are executed")
```

````python
#wirte a program to find the vowels from sting using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if (char in "aeiuoAEIOU"):
        print(char)
````

````python
#write a program to find the non vowels from sting using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char not in "aeiouAEIOU"):
        print(char)
````
